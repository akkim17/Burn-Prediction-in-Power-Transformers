```python
data1.rename(columns = {'Burned transformers 2019':'Burned transformers'}, inplace = True)
data2.rename(columns = {'Burned transformers 2020':'Burned transformers'}, inplace = True)
```

```python
merged_data=pd.concat([data1,data2],axis=0)
```

```python
merged_data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
for i in merged_data.columns:
    if merged_data[i].dtype=='object':
        merged_data[i]=enc.fit_transform(merged_data[i])
merged_data.info()
```

```python
merged_data.drop_duplicates()
merged_data.info()
```

```python
merged_data.describe()
```

```python
y=merged_data['Burned transformers']
x=merged_data.drop(['Burned transformers'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
x=std.fit_transform(x)
```
